```python
import nltk.chunk
 
def conll_tag_chunks(chunk_sents):
    tag_sents = [nltk.chunk.tree2conlltags(tree) for tree in chunk_sents]
    return [[((w,t),c) for (w, t, c) in chunk_tags] for chunk_tags in tag_sents]
```
