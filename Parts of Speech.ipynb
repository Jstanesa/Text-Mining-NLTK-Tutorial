This section of the tutorial will teach you how to tag words with their parts of speech. First, we are going to open up our saved text that we retrieved earlier in the tutorial.

```python
import nltk
```

```python
f = open("255s.txt","rU") #U is for Universal
raw = f.read()
f.close()
```

Now, we will break up our text into sentences, and then words. After we do this, we tag all the words with their corresponding parts of speech.

```python
sentences = nltk.sent_tokenize(raw) #create list of sentences
sentences = [nltk.word_tokenize(sent) for sent in sentences] #convert each sentence into a list of words
sentences = [nltk.pos_tag(sent) for sent in sentences] #for each sentence, tag each word
```

This leaves us with a nested array of (word, tag) pairs.
